## Project Aims
This project will explore a simple application of the empirical Bayes method and compare it to typical frequentist A/B testing methods.

The project simulates an experiment that compares a number of test subjects (typically more than two) where each subject can be used interchangeably and the success or failure of the subject is an independant Bernoulli trial with parameter p. For example each subject could be a headline on an article link.

In this simulation the true value of p for each subject is known and the purpose is to see how effective the empirical Bayes method is at identifying the subject with the greatest true p parameter, that is, the best test subject.

## Assumptions and Limitations
As stated above each trial is considered to be an independant Bernoulli trial. Furthermore it is assumed that the p parameter of each subject is constant. For example the probability of a recipient clicking on the article link is the same for the entire population of potential recipients.

It is also assumed that the result of each trial is observed before the next trial is conducted. In practice this may difficult for some applications and it might be neccesary to make decisions based on the current but incomplete data. Also a rule to determine when success or failure is recorded would be required.

To avoid excessively long computation times I will limit the number of tests run in any investigation of the method. Running the project on my laptop the practical limit is around 4000 - 5000 tests depending on the parameter values.

## Methods
This project is based on multi-armed bandit (MAB) techniques. Here the aim is primarily exploration as normally defined in MAB problems, that is to find the subject with the greatest p parameter. The exploitation aspect of the method aims to collect more data on the more promising subjects and less on the less promising subjects.

Each test subject is initially assigned a prior Beta distribution. The distribution parameters can be intrepreted as the number of successes and the number of failures. The distribution parameters are updated after each trial to produce a posterior distribution.

For each trial a random value is generated from each subject's distribution and the subject corresponding to the greatest random value is selected for the trial. Hence the method will tend to select the subjects with the greater posterior means / estimated p parameter values more frequently. 

In order to identify an optimal subject this application eliminates test subjects when the likelihood of them being the optimal subject is considered sufficiently low. The actual probability of each subject being optimal is either 1 or 0 so calculating an accurate probability is not possible without complete information. Instead the method I have chosen uses an elimination metric as a proxy for the probability of a subject being optimal. As data is collected the elimination metric is updated in an effort to reach the true probability for each subject, that is 0 or 1. The elimination metric is a function of the current posterior mean and standard deviation and expressed as a proportion. For each subject the elimination metric is given by (posterior mean / standard deviation) as a proportion of the sum of all (posterior mean / standard deviation) values. Hence subjects with greater posterior means will generally have greater elimination metrics. Also lower standard deviations will result in greater elimination metric values.

## Questions
The empirical bayes method employed in this project has four parameters. Two of these, the initial values of n_succ and n_fail, describe the prior knowledge of the each subject. The other two, threshold and max_trials, define how the experiment is run. One aim of this project is to investigate how changing these parameters affects the performance of the method.

Typically a frequentist A/B test is designed to have a particular significance level and power for a given conversion rate (the p parameter in the Bernoulli trials in this case) and percentage change to the conversion rate that you wish to detect. Having specified these requirements a sample size can be calculated. This project will compare the number of trials required to reach a result using this method against the sample size required for the frequentist method.

As a baseline for comparison I will use the following:<br>
    Existing / control conversion rate: 0.1<br>
    Minimum detectable effect: 20% (one-sided test, assuming treatment will be better than control)<br>
    Significance level: 0.05<br>
    Power: 0.8<br>
    
Online sample size calculators give a variety of values for the above specification, ranging from 5600 to 7200 total for both control and treatment.

In comparing this method against a frequentist A/B test method the probability of a false positive or false negative result are two metrics of interest.

```python
# install modules
import pandas as pd
import numpy as np
import random
from matplotlib import pyplot as plt
import seaborn as sns
from operator import attrgetter
```

```python
class Test_Subject:
    
    """represents a prototype to be tested
    
        parameters
        ----------
        name: str
            instance name
        true_p: float
            true value of p value / success rate
        n_succ: integer
            number of successes / a parameter of Beta distribution
        n_fail: integer
            number of failures / b parameter of Beta distribution
        elim_status: boolean
            True if subject has been eliminated from test
        est_p: float
            current estimate of p / mean of posterior Beta distribution
        est_sd: float
            standard deviation of posterior Beta distribution
        elim_metric: float
            metric to determine if subject should be eliminated from test based on observed results
    """
    # define an init method
    def __init__(self, name, true_p, n_succ, n_fail):
        self.name = name
        self.true_p = true_p
        self.n_succ = n_succ
        self.n_fail = n_fail
        self.elim_status = False
        self.est_p = self.n_succ / (self.n_succ + self.n_fail)
        self.est_sd = (self.n_succ * self.n_fail) / ((self.n_succ + self.n_fail)**2 * (self.n_succ + self.n_fail + 1))
        self.elim_metric = 0
        
    # define a repr method
    def __repr__(self):
        return self.name
    
    # define a calc est_p method
    def calc_est_p(self):
        self.est_p = self.n_succ / (self.n_succ + self.n_fail)
        
    # define a calc_est_sd
    def calc_est_sd(self):
        self.est_sd = (self.n_succ * self.n_fail) / ((self.n_succ + self.n_fail)**2 * (self.n_succ + self.n_fail + 1))
```

```python
def run_test(subject_dict, threshold_tune, max_trials):
    
    """function to define and run the desired test
    
        parameters
        ----------
        subject_dict: dictionary object
            dictionary of the subjects to be tested
        threshold_tune: float
            parameter to fine tune the threshold for subject elimination
        max_trials: interger
            the maximum number of trials
    """
    
    # set up data structures required
    
    ## create results dataframe
    ### create list of column names with trial_no column
    col_list = ['trial_no']

    ### add column names for each subject
    for subject in subject_dict:
        
        col_list.append(f'{subject}_elim_metric')
        col_list.append(f'{subject}_n_succ')
        col_list.append(f'{subject}_n_fail')
        col_list.append(f'{subject}_est_p')
    
    ### create variable to sum initial est_p/est_sd ratios
    p_sd_ratio_sum = 0
    
    ### create empty list of subject data for first row of dataframe
    data = []

    ### sum initial p_sd_ratio for each subject
    for subject in subject_dict:
    
        p_sd_ratio_sum += (subject.est_p / subject.est_sd)

    ### add initial values to data list for each subject
    for subject in subject_dict:
    
        data.append((subject.est_p / subject.est_sd) / p_sd_ratio_sum)
        data.append(subject.n_succ)
        data.append(subject.n_fail)
        data.append(subject.est_p)
    
    ### combine subject data and initial trial_no
    df_data = [[0] + data]
    
    ### create result_df
    result_df = pd.DataFrame(data = df_data, columns=col_list)
    
    ## create a total trial counter
    tot_trials = 0
    
    ## create a subject counter
    no_subject = len(subject_dict)
    
    ## create threshold
    threshold = (1 / (no_subject + 1)) + (threshold_tune * no_subject**2)
    
    # iterate until the optimal subject is identified or the maximum number of trials is reached
    while (tot_trials < max_trials) & (no_subject > 1):
        
        #### alternate version
        ## generate random sample from current subject posterior distribution
        
        ## loop through subjects in subject_dict
        for subject in subject_dict:
            
            ### include only subjects that have not been eliminated
            if subject.elim_status == False:
            
                ### sample from posterior beta distribution for each subject
                subject_dict[subject] = random.betavariate(subject.n_succ, subject.n_fail)
                                  
    
        ## select the subject with the greatest estimated mean / random variable
        selection = max(subject_dict, key=subject_dict.get)
        
        #### alternate version
        
        ## check whether the trial is successful
        if selection.true_p > random.random():
            
            ### update the subject n_succ
            selection.n_succ += 1
            
            ### update the subject est_p attribute
            selection.calc_est_p()
            
            ### update the subject est_sd attribute
            selection.calc_est_sd()
            
        else:
            
            ### update the subject n_fail
            selection.n_fail += 1
    
            ### update the subject est_p attribute
            selection.calc_est_p()
            
            ### update the subject est_sd attribute
            selection.calc_est_sd()
    
        # update elim_metric values
        ## reset p_sd_rato_sum
        p_sd_ratio_sum = 0
        
        ### sum p_sd ratios for each subject
        for subject in subject_dict:
            
            ### include only subjects that have not been eliminated
            if subject.elim_status == False:
    
                p_sd_ratio_sum += (subject.est_p / subject.est_sd)
        
        ### calculate elim_metric for each subject
        for subject in subject_dict:
            
            ### include only subjects that have not been eliminated
            if subject.elim_status == False:
            
                subject.elim_metric = ((subject.est_p / subject.est_sd) / p_sd_ratio_sum)
        
        # update tot_trials
        tot_trials += 1

        # save iteration outcome to dataframe
        
        ## create empty list for subject trial data
        trial_data = []
    
        ## add trial values to data list for each subject
        for subject in subject_dict:
    
            trial_data.append(subject.elim_metric)
            trial_data.append(subject.n_succ)
            trial_data.append(subject.n_fail)
            trial_data.append(subject.est_p)
        
        ## combine subject data and current trial_no
        trial_df_data = [[tot_trials] + trial_data]
    
        ## add trial data to result_df
        result_df = pd.concat([result_df, pd.DataFrame(trial_df_data, columns=result_df.columns)], ignore_index=True)
    
        # check if threshold for elimination has been passed
        
        ## find subject with minimum elim_metric that has not been eliminated
        min_subject = min([x for x in subject_dict if x.elim_status == False], key=attrgetter('elim_metric'))
        
        #print(f'The subject with minimum elim_metric is {min_subject} with elim_metric of {min_subject.elim_metric}')
        
        if min_subject.elim_metric < threshold:
            
            #print(f'The subject eliminated is {min_subject} with elim_metric of {min_subject.elim_metric}')
            
            ### set subject elim_status to True
            min_subject.elim_status = True
            
            ### set subject elim_metric to zero
            min_subject.elim_metric = 0
            
            ### set subject_dict[subject] to zero
            subject_dict[min_subject] = 0
            
            ### update no_subjects
            no_subject -= 1
                
            ### update threshold
            threshold = (1 / (no_subject + 1)) + (threshold_tune * no_subject**2)
                
    # confirm if an optimal subject has not been identified, save it if appropriate
    test_result = 0
    opt_subject = None
    p_error = None
    
    # check if an optimal subject has been identified
    if no_subject == 1:
        
        ## update test_success variable
        test_result = 1
        
        for subject in subject_dict:
            
            if subject.elim_status == False:
                
                opt_subject = subject
                
                break
                
        ## calculate the est_p error for the opt_subject
        p_error = ((opt_subject.est_p - opt_subject.true_p) / opt_subject.true_p) * 100

    # return the remaining subjects (the best subject) and the results_df
    return test_result, opt_subject, p_error, tot_trials, result_df
```

There are two potential problems with the process of eliminating subjects from the test; the true optimal subject could be eliminated due to a string of unrepresentative results causing the elimination metric to pass the elimination threshold, or a non-optimal subject might not be eliminated before the maximum number of trials is reached because the elimination metric remains above the elimination threshold. 

The first problem suggests setting a low elimination threshold whereas the second problem suggests setting a high elimination threshold. In order to address these contradictory pressures the threshold for elimination is dynamic, it increases as the number of subjects is reduced. This allows a low threshold initially reducing the risk of an optimal subject being eliminated early in the test. As the number of subjects is reduced the threshold increases reducing the risk that the test does not produce a result.

To allow for fine tuning of the elimination threshold the formula includes a threshold_tune parameter. Positive values of threshold_tune will increase the threshold, negative values will decrease the threshold. I illustrate how the elimination threshold varies with the number of subjects and values of threshold_tune below. 

```python
# investigate threhold values as threshold_tune varies

## create list of possible threshold_tune values
threshold_tune_vals = [x * 0.001 for x in range(-3,4)]

## set maximum number of subjects
max_subjects = 4

## create a dictionary to store threshold values
threshold_dict = {}

## calculate threshold values
for threshold_tune in threshold_tune_vals:
    
    threshold_list = []
    
    for no_subject in range(2,max_subjects+1):

        threshold_list.append((1 / (no_subject + 1)) + (threshold_tune * no_subject**2))
        
    threshold_dict[threshold_tune] = threshold_list
    
# plot threshold values

## set x axis values
x_values = list(range(2,max_subjects+1))

## create figure, axes
fig, ax = plt.subplots(figsize=(6,4))

## plot line for each threshold_tune value
for threshold_tune in threshold_tune_vals:
    
    plt.plot(x_values, threshold_dict[threshold_tune], label=threshold_tune)
    
## modify axes & add legend
ax.set_xticks(x_values)
ax.invert_xaxis()
plt.legend()

## format plot
plt.title('Elimination Metric vs Number of Subjects')
plt.xlabel('No. Subjects')
plt.ylabel('Elimination Metric')

## display plot
plt.show()  
```

```python
def plot_result(result_df, subject_dict, file_name=False):

    """function to plot the results of a test
    
        parameters
        ----------
        result_df: DataFrame object
            subject attribute values throughout test
        subject_dict: dictionary object
            dictionary of the subjects to be tested
    """
    
    # create lists of columns to plot
    elim_metric_cols = []
    est_p_cols = []
    
    # create a list of colours (assuming a max of 4 subjects, first is the control)
    colour_list = ['firebrick', 'darkgreen', 'limegreen', 'lawngreen']

    # loop through columns and add to lists
    for col in result_df.columns:
    
        # check if column is _prop column
        if '_elim_metric' in col:
        
            elim_metric_cols.append(col)
        
        if '_est_p' in col:
        
            est_p_cols.append(col)

    # create a figure
    plt.figure(figsize=(15,6))
    ## give it a title
    plt.suptitle('Experiment Results')

    # plot the _prop columns
    ax1 = plt.subplot(1,2,1)
    
    ## loop through _prop columns
    for i in range(len(elim_metric_cols)):
        
        ### plot prop vs trial
        plt.plot(result_df['trial_no'], result_df[elim_metric_cols[i]], label=elim_metric_cols[i], c=colour_list[i])
        
    ## format plot
    ax1.spines[['top', 'right']].set_visible(False)
    ax1.set_axisbelow(True)
    ax1.yaxis.grid(True, color='#DDDDDD')
    plt.title('Elimination Metric vs Trial')
    plt.xlabel('Trial')
    plt.ylabel('Elimination Metric')
    plt.legend()
    
    # plot the _est_p columns
    ax2 = plt.subplot(1,2,2)
    
    ## loop through _est_p columns
    for i in range(len(est_p_cols)):
        
        ### plot est_p vs trial
        plt.plot(result_df['trial_no'], result_df[est_p_cols[i]], label=est_p_cols[i], c=colour_list[i])
        
        ### plot true_p
        plt.axhline(y=list(subject_dict.keys())[i].true_p, ls='--', c=colour_list[i])
    
    ## format plot
    ax2.spines[['top', 'right']].set_visible(False)
    plt.title('Estimated p Parameter vs Trial')
    plt.xlabel('Trial')
    plt.ylabel('Estimated p Parameter')
    plt.legend()
    
    # save the figure
    if file_name != False:
        
        plt.savefig(file_name)
    
    # return the figure
    plt.show()
```

```python
# create function to repeat a test
def repeat_test(subject_dict, threshold_tune, max_trials, no_test):
    
    """function to define and run the desired test repeatedly
    
        parameters
        ----------
        subject_dict: dictionary object
            dictionary of the subjects: proportions to be tested
        threshold_tune: float
            tuning value for subject elimination threshold
        max_trials: interger
            the maximum number of trials in a single test
        no_test: integer
            the number of times to repeat the test
    """
    
    # create data structures to store results
    
    ## create performance_df
    
    ### columns are test_no, test_result, test_opt_subject, test_success, tot_trials
    ### test_no: number of test iteration
    ### test_result: 1 if an optimal subject was identified, 0 otherwise
    ### test_opt_subject: the subject identified as the optimal subject, None otherwise
    ### p_error: the percentage error of the est_p for the subject identified as the optimal subject
    ### test_success: 1 if the true optimal subject was identified, 0 otherwise
    ### tot_trials: number of trials required to identify an optimal subject
    
    ### create column list
    col_list = ['test_no', 'test_result', 'test_opt_subject', 'p_error', 'test_success', 'tot_trials']
    
    ### create dataframe
    performance_df = pd.DataFrame(columns = col_list)
    
    # run test no_test times
    
    ## set test_counter
    test_counter = 0
    
    ## set true optimal subject
    true_opt = max(subject_dict, key=attrgetter('true_p'))
    
    ## set initial values
    init_value_dict = {}
    
    for subject in subject_dict:
        init_value_dict[subject] = [subject.n_succ, subject.n_fail]
        
        ### save initial n_succ and n_fail values (assumes all subjects have same values)
        succ_fail = f'{subject.n_succ}/{subject.n_fail}'
    
    
    ## repeat test until test_counter = no_test
    while test_counter < no_test:
        test_result, opt_subject, p_error, tot_trials, test_df = run_test(subject_dict=subject_dict, \
                                                                     threshold_tune=threshold_tune, max_trials=max_trials)
    
        ### check is true optimal subject was identified
        if opt_subject == true_opt:
            
            test_success = 1
            
        else:
            
            test_success = 0
        
        ### store opt_subject name or None
        if type(opt_subject) == Test_Subject:
            
            opt_subject_name = opt_subject.name
            
        else:
            
            opt_subject_name = None
        
        opt_subject_name
        
        ### save results to list
        test_data = [[test_counter, test_result, opt_subject_name, p_error, test_success, tot_trials]]
    
        ### save results to performance_df
        performance_df = pd.concat([performance_df, pd.DataFrame(data=test_data, columns=col_list)], ignore_index=True)
        
        ### update test_counter
        test_counter += 1
        
        ### reset subject attribute values for next test
        for subject in subject_dict:
            
            subject.n_succ = init_value_dict[subject][0]
            subject.n_fail = init_value_dict[subject][1]
            subject.elim_status = False
            subject_elim_metric = 0
            subject.calc_est_p()
            subject.calc_est_sd()
    
    # calculate and store performance indicators
    # performance indicators are:
    # percentage of tests that produced an optimal subject (true or false)
    # percentage of tests that identified the true optimal subject
    # average p_error
    # average number of trials required to produce result
    # average number of trials required to produce correct result
    # average number of trials for all tests

    # calculate performance indicators
    perc_result = round(performance_df['test_result'].mean(), 3) * 100
    perc_success = round(performance_df['test_success'].mean(), 3) * 100
    avg_p_error = round(performance_df['p_error'].mean(), 2)
    avg_result_trial = round(performance_df[performance_df['test_result'] == 1]['tot_trials'].mean(), 1)
    avg_success_trial = round(performance_df[performance_df['test_success'] == 1]['tot_trials'].mean(), 1)
    avg_all_trial = round(performance_df['tot_trials'].mean(), 1)

    # create list with test parameters and performance indicators
    perf_ind_data = [[threshold_tune, max_trials, succ_fail, perc_result, \
                      avg_result_trial, avg_p_error, perc_success, avg_success_trial, avg_all_trial]]

    # create list of column headers
    perf_ind_cols = ['Threshold Tune Value', 'Max Trials', 'Initial Succ/Fail', 'Test Result %', \
                     'Avg Result Trials', 'Avg est_p Error %', 'Test Success %', 'Avg Success Trials', 'Avg All Trials']

    # create perf_ind_df
    perf_ind_df = pd.DataFrame(data=perf_ind_data, columns=perf_ind_cols)
    
    # return performance_df
    return performance_df, perf_ind_df
```

```python
# create function to run repeat_test with different parameters
## possible parameters are:
## threshold_tune, for example 0.000, 0.001, 0.002, 0.003, 0.004
## max_trials, for example 1000, 2000, 3000, 4000, 5000
## initial n_succ and n_fail values, for example 10 & 90 or 20 & 180

def parameter_search(subject_dict, threshold_tune_vals, max_trials_vals, para_multipliers, no_test):
    
    """function to define and run the desired test repeatedly with a range of parameter values
    
        parameters
        ----------
        subject_dict: dictionary object
            dictionary of the subjects: proportions to be tested
        threshold_tune_vals: list object
            list of threshold_tune values to be used
        max_trials_vals: list object
            list of the max_trials values to be used
        para_multipliers: list object
            list of values to multiply baseline subject n_succ and n_fail attributes
        no_test: integer
            the number of times to repeat the test
    """
    
    ### store baseline distribution parameter values
    base_dist_para_dict = {}
    
    for subject in subject_dict:
        base_dist_para_dict[subject] = [subject.n_succ, subject.n_fail]

    # create perf_ind_comp_df
    # create list of column headers
    perf_ind_cols = ['Threshold Tune Value', 'Max Trials', 'Initial Succ/Fail', 'Test Result %', \
                     'Avg Result Trials', 'Avg est_p Error %', 'Test Success %', 'Avg Success Trials', 'Avg All Trials']

    # create perf_ind_comp_df
    perf_ind_comp_df = pd.DataFrame(columns=perf_ind_cols)
    
    # loop through para_multipliers values
    for para_multiplier in para_multipliers:
        
        # loop through subjects in subject_dict
        for subject in subject_dict:
            
            ## set initial n_succ and n_fail
            subject.n_succ = subject.n_succ * para_multiplier
            subject.n_fail = subject.n_fail * para_multiplier
    
        # loop through threshold_tune values
        for threshold_tune in threshold_tune_vals:
        
            ## loop through max_trials values
            for max_trials in max_trials_vals:
            
                ### run tests
                performance_df, perf_ind_df = repeat_test(subject_dict=subject_dict, \
                                                      threshold_tune=threshold_tune, max_trials=max_trials, no_test=no_test)
            
                ### save results to perf_ind_comp_df
                perf_ind_comp_df = pd.concat([perf_ind_comp_df, perf_ind_df], ignore_index=True)
                
        ### reset distribution parameter values
        for subject in subject_dict:
                    
            #### reset n_succ and n_fail to baseline
            subject.n_succ = base_dist_para_dict[subject][0]
            subject.n_fail = base_dist_para_dict[subject][1]
            
    return perf_ind_comp_df 
```

```python
# visualise the parameter search results
def plot_performance(perf_ind_comp_df, para_int):

    """function to plot parameter performance
    
        parameters
        ----------
        perf_ind_comp_df: DataFrame object
            subject attribute values throughout test
        para_int: string
            perf_ind_comp_df column header for parameter of interest
    """
    
    # set x_vals
    x_vals = np.arange(len(perf_ind_comp_df[para_int]))
    
    # set bar_width
    bar_width = 0.4
    
    # create a figure
    plt.figure(figsize=(15,10))
    ## give it a title
    plt.suptitle(f'{para_int} Parameter Performance')

    # plot the test result % and test success % on first subplot
    ax1 = plt.subplot(2,2,1)
    
    ## plot test result % against parameter of interest
    plt.bar(x_vals, perf_ind_comp_df['Test Result %'], width=bar_width, label='Test Result %')
    
    ## plot test success % against parameter of interest
    plt.bar(x_vals + bar_width, perf_ind_comp_df['Test Success %'], width=bar_width, label='Test Success %')
        
    plt.title(f'Test Result % and Test Success % vs {para_int}')
    ax1.spines[['top', 'left', 'right']].set_visible(False)
    ax1.set_axisbelow(True)
    ax1.yaxis.grid(True, color='#DDDDDD')
    plt.xlabel(f'{para_int}')
    plt.ylabel('%')
    plt.xticks(x_vals + (bar_width / 2), perf_ind_comp_df[para_int])
    plt.legend()
    
    # plot test success as percentage of test results
    ax2 = plt.subplot(2,2,2)
    
    ## calculate the y values
    succ_result_vals = (perf_ind_comp_df['Test Success %'] / perf_ind_comp_df['Test Result %']) * 100
    
    ## plot test success % against parameter of interest
    plt.bar(x_vals, succ_result_vals)
    
    ax2.spines[['top', 'left', 'right']].set_visible(False)
    ax2.set_axisbelow(True)
    ax2.yaxis.grid(True, color='#DDDDDD')
    ax2.set_ylim(bottom=50, top=100)
    plt.title(f'Test Success as % of Test Result vs {para_int}')
    plt.xlabel(f'{para_int}')
    plt.ylabel('%')
    plt.xticks(x_vals, perf_ind_comp_df[para_int])
    
    # plot the test result, test success and all tests average number of trials on third subplot
    ax3 = plt.subplot(2,2,3)
    
    ## plot test result average number of trials against parameter of interest
    plt.bar(x_vals*1.4, perf_ind_comp_df['Avg Result Trials'], width=bar_width, label='Avg Trials for Result')
    
    ## plot avg trials for success against parameter of interest
    plt.bar(x_vals*1.4 + bar_width, perf_ind_comp_df['Avg Success Trials'], width=bar_width, label='Avg Trials for Success')
    
    ## plot avg trials for all tests against parameter of interest
    plt.bar(x_vals*1.4 + 2*bar_width, perf_ind_comp_df['Avg All Trials'], width=bar_width, label='Avg Trials for all Tests')
    
    ax3.spines[['top', 'left', 'right']].set_visible(False)
    ax3.set_axisbelow(True)
    ax3.yaxis.grid(True, color='#DDDDDD')
    plt.title(f'Avg Trials vs {para_int}')
    plt.xlabel(f'{para_int}')
    plt.ylabel('Avg Trials')
    plt.xticks(x_vals*1.4 + bar_width ,perf_ind_comp_df[para_int])
    plt.legend()
    
    # plot est_p error % on second subplot
    ax4 = plt.subplot(2,2,4)
    
    ## plot est_p error % against parameter of interest
    plt.bar(x_vals, perf_ind_comp_df['Avg est_p Error %'])
        
    ax4.spines[['top', 'left', 'right']].set_visible(False)
    ax4.set_axisbelow(True)
    ax4.yaxis.grid(True, color='#DDDDDD')
    plt.title(f'Avg est_p Error % vs {para_int}')
    plt.xlabel(f'{para_int}')
    plt.ylabel('%')
    plt.xticks(x_vals ,perf_ind_comp_df[para_int])
    
    # return the figure
    plt.show()
```

## Conclusions

Unlike frequentist A/B testing this method's performance is not quantified by mathematical theory, instead these conclusions rely on simulated results. For greater confidence in the results very large numbers of simulations would ideally be done, 10,000 or more. Unfortunately the computing time required has limited the number I could conveniently perform. However the results do appear to be reasonably predictable and consistent suggesting random variation has not affected them excessively.

### Power of test
We have seen that the method can be modified to achieve a high true positive rate or power of test. In this demonstation I have selected parameter values to maximise the performance of the model in the situation that there is a treatment 20% better than the control. In those circumstances the method has approximately 88% power. Where the actual difference between the control and treatment 1 is only 10% the power of the test dropped to approximately 60%.

### Significance level of test
Unlike frequentist A/B testing this method does not perform any significance testing to measure the probability that the result observed is due to random variation and not a genuine difference in p parameter values. Instead this method tries to identify the subject with the highest p parameter and as illustrated in the last experiment, when two subjects have the same or very similar p parameters the method will identify the two subjects as optimal with approximate equal frequency. So compared to frequentist testing a large number of false positives will be seen. The upside is that the method tends to provide reasonable (within 5%) estimates of the true p parameter. Hence if a treatment is identified as the optimal subject with an estimated p parameter of around 0.1 +/- 5% when the control is known to have a p parameter of 0.1 it is clear that there is not strong evidence that the treatment is substantially better than the control.

In cases where the control was substantially better than any treatment, say 10 or 20% better, we would expect results similar to the experiments above except with the labels of control and treatment 1 swapped. In those experiments we saw 2.6% and 9.9% of the tests incorrectly identify the control as optimal when treatment 1 was 20% and 10% better respectively.

If avoiding false positives was considered to be very important the method could be modified to reduce the false positive rate. Choosing the threshold tune parameter or the initial successes/failures parameter so that more data is required for a subject to be eliminated will reduce the false positive rate and increase the number of tests not producing a result. Alternatively the elimination threshold for only the control could be raised so that the method will require a greater level of evidence to eliminate the control compared to the treatment subjects.

### Sample size
Where there was a 20% difference between the control and the best treatment an average of approximately 2750 trials were required to identify the true optimal subject. Approximately 9% of tests ran to the maximum number of trials, in this case 6000 trials. The average number of trials required over all tests was 3037.

When the difference between the control and the best treatment was only 10% 3170 trials were required to identify the true optimal subject. Almost 30% of tests ran to the maximum number of trials. The average number of trials required over all tests was 4008.

In the situation where the best treatment is 1% worse than the control 48.5% of tests ran to the maximum numer of trials and the average number of trials over all tests was 4718.

We can see that the method requires less trials when there are greater differences between the subject p parameters. As stated in the introduction sample size calculators for frequentist A/B tests recommend between 5600 and 7200 trials split equally between the two subjects. So in all cases the average number of trials required is less than the frequentist A/B tests but some tests will run to the maximum 6000 trials. The proportion of tests requiring 6000 trials will increase as the difference between the subjects reduces.

### Conclusion
Benefits:<br>
- Allows more than two subjects to be tested.<br>
- High power can be achieved when there is a 20% difference in p parameters.<br>
- The more promising subjects are selected for trials more often, reducing the number of trials using the less promising subjects.<br>
- Less trials required on average than a comparable frequentist test.<br>
- Produces results with less trials when there is a greater difference between subjects.<br>
- Parameters can be adjusted to optimize results depending on circumstances and priorities.<br>

Drawbacks:<br>
- Number of trials to completion is unknown, up to the maximum specified.<br>
- Where the control and a treatment (or two treatments) have similar p parameters a high number of false positives will occur.<br>
- The subject distributions need to be updated during the test.<br>

This method is preferable to the frequentist A/B approach when there are more than 2 subjects, where faster results are required, where the solution is applicable for only a short time and you wish to exploit the results during the test by eliminating the worst subjects quickly.

This method is not appropriate where a false positive result would have serious implications in cost or time, where a defined end point is required, where updating the subject distributions during the test would be impractical. All these issues could be addressed to some degree by modifications to the method.

Overall the method performed well compared to a frequentist A/B test in the situation where there is a 20% difference between the control and the best treatment. For a smaller sample size (average 3037), greater power (88% versus 80%) was achieved.
